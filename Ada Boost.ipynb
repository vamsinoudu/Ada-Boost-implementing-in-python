```python
df.drop('RowNumber',inplace=True,axis=1)
```

```python
df.drop('CustomerId',inplace=True,axis=1)
```

```python
df.drop('Surname',inplace=True,axis=1)
```

```python
# Get independent and dependent features :

x=df.iloc[:,0:10]
y=df.iloc[:,10]
```

```python
geography=pd.get_dummies(x['Geography'],drop_first=True)
```

```python
geography.head()
```

```python
gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
gender.head()
```

```python
# Drop categorical features:

x=x.drop(['Geography','Gender'],axis=1)
```

```python
x.head()
```

```python
x=pd.concat([x,geography,gender],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
Adamodel=AdaBoostClassifier(n_estimators=100,learning_rate=1)
```

```python
model=Adamodel.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# model Accuracy:

print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```

```python
# Use the logistic regression:

from sklearn.linear_model import LogisticRegression
mylogsticreg=LogisticRegression()
```

```python
abcd=AdaBoostClassifier(n_estimators=200,base_estimator=mylogsticreg,learning_rate=1)
```

```python
model=abcd.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```
